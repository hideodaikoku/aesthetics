```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_csv('../output/IRR.csv', sep=',',header=None)
```

```python
df
```

```python
labels = pd.read_csv('../output/labels.csv', sep=',')
```

```python
labels=labels.values
```

```python
labels = ['Annotator 1','Annotator 2', 'Annotator 3','Annotator 4','Annotator 5','Annotator Avg.','Algorithm 1 ','Algorithm 2 ','Algorithm Average']
```

```python
df
```

```python
matplotlib.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(figsize=(34.4,15))
colormap = plt.cm.get_cmap('BuPu_r')
ax = sns.heatmap(df, annot=True, linewidths=.5,cmap ="PuBu")
mask = np.zeros_like(df)
mask[np.tril_indices_from(mask)] = True
with sns.axes_style("white"):
    asx = sns.heatmap(df, mask=mask,annot=True, linewidths=.5, cmap =colormap)
asx.set_yticklabels(labels, rotation =0)
asx.set_xticklabels(labels, rotation =60)
```
